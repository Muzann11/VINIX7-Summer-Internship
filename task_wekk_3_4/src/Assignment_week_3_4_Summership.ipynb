```python
df_desc.dropna(inplace=True)
df_desc.drop_duplicates(inplace=True)
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt_tab')    # untuk word_tokenize()
nltk.download('stopwords')    # untuk stopwords
nltk.download('wordnet')      # untuk lemmatizer
nltk.download('omw-1.4')      # untuk lemmatizer support

stop_words = set(stopwords.words('english'))
```

```python
def to_lowercase(text):
    return text.lower()

def normalize_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()

def remove_urls(text):
    return re.sub(r'http\S+|www\.\S+', '', text)

def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)

def remove_emails(text):
    return re.sub(r'\S+@\S+', '', text)

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_stopwords(text):
    words = word_tokenize(text)
    filtered = [word for word in words if word not in stop_words]
    return ' '.join(filtered)

# Aplikasikan semua preprocessing
df_desc = (
    df_desc.apply(to_lowercase)
           .apply(remove_urls)
           .apply(remove_html_tags)
           .apply(remove_emails)
           .apply(remove_numbers)
           .apply(remove_punctuation)
           .apply(normalize_whitespace)
           .apply(remove_stopwords)
)
```

```python
all_cleaned_text = " ".join(df_desc.astype(str).tolist())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Inisialisasi vectorizer
vectorizer = TfidfVectorizer(max_df=0.85, max_features=1000)

# Fit dan transform
tfidf_matrix = vectorizer.fit_transform(df_desc)
feature_names = vectorizer.get_feature_names_out()

# Fungsi untuk ekstrak top-N keyword per dokumen
def extract_top_keywords(tfidf_row, feature_names, top_n=5):
    sorted_nzs = tfidf_row.nonzero()[1][tfidf_row.data.argsort()[::-1]]  # sort descending
    top_features = [(feature_names[i], tfidf_row[0, i]) for i in sorted_nzs[:top_n]]
    return [word for word, score in top_features]

# Terapkan pada setiap baris
top_keywords = []
for i in range(tfidf_matrix.shape[0]):
    keywords = extract_top_keywords(tfidf_matrix[i], feature_names, top_n=5)
    top_keywords.append(keywords)

# Simpan hasil ke DataFrame
df_keywords = pd.DataFrame({
    'original_desc': df_desc,
    'top_keywords': top_keywords
})

df_keywords.head()
```

```python
import numpy as np

# Jumlahkan semua skor tf-idf tiap fitur/kolom (menjadi total skor untuk masing-masing kata)
sums = tfidf_matrix.sum(axis=0)  # shape: (1, n_features)
scores = np.asarray(sums).flatten()

# Buat DataFrame untuk memetakan kata dan skornya
keywords_scores = pd.DataFrame({
    'word': feature_names,
    'score': scores
}).sort_values(by='score', ascending=False)

keywords_scores.head(20)  # top 20 keyword di seluruh dataset
```
