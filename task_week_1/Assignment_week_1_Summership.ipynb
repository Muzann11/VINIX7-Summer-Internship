```python
df = data.copy()

df.replace("...", pd.NA, inplace=True)
```

```python
baris_error = df[df.isna().any(axis=1)]
baris_error
```

```python
df = df.drop(baris_error.index)
df
```

```python
df.shape
```

```python
df = df[df['Provinsi'] != 'Indonesia']

df['Provinsi'] = df['Provinsi'].astype(str)

for col in df.columns:
  if col != 'Provinsi':
    df[col] = df[col].astype(float)

df = df.reset_index(drop=True)

df.info()
```
